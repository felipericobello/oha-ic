```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def solution(w0, Beta, Amplitude, Size, TimeInterval):
    w1 = np.sqrt(pow(w0, 2) - pow(Beta, 2))
    
    #x(t) = A*np.exp(-Beta*time)*np.cos(w1*t + phase)
    
    x = []
    tempo = []
    for iter in range(Size):
        t = (TimeInterval * iter)
        tempo.append(t)
        xt = Amplitude * np.exp(-Beta*t) * np.cos(w1*t)
        x.append(xt)

    x = np.array(x, dtype=np.float16)
    t = np.array(t, dtype=np.float16)

    return x, tempo

def derivative_solution(w0, Beta, Amplitude, Size, TimeInterval):
    w1 = np.sqrt(pow(w0, 2) - pow(Beta, 2))
    
    #x(t) = A*np.exp(-Beta*time)*np.cos(w1*t + phase)
    
    x = []
    tempo = []
    for iter in range(Size):
        t = (TimeInterval * iter)
        tempo.append(t)
        xt = -Amplitude * np.exp(-Beta*t) * (Beta*np.cos(w1*t) + w1*np.sin(w1*t))
        x.append(xt)
    x = np.array(x, dtype=np.float16)
    t = np.array(t, dtype=np.float16)

    return x, tempo
```

```python
W0   = 1.80
BETA = 0.15
A    = 1.00

time_units = 10
DT   = 0.01
SIZE =  (int)(time_units/DT)


x, t = solution(W0, BETA, A, SIZE, DT)
xl, t = derivative_solution(W0, BETA, A, SIZE, DT)

plt.plot(t, x, 'b-')
plt.title('Função OHA x(t), D(x) = {0, 10}')
plt.xlabel('Tempo [unidade de tempo]')
plt.ylabel('Posição [unidade de pos.]')
plt.grid(True, alpha=0.3)
plt.savefig('grafico-funcao-xt.jpg', dpi=150)
plt.show()

plt.plot(x, xl, 'k-')
plt.title('Diagrama de fase, D(x) = {0, 10}')
plt.xlabel('x(t) [unidade de pos.]')
plt.ylabel('dx/dt(t) [unidade de pos.]/[unidade de tempo]')
plt.grid(True, alpha=0.3)
plt.savefig('grafico-diagrama-de-fase.jpg', dpi=150)
plt.show()
```
